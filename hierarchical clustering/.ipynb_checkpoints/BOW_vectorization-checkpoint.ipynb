```python
#corpus of docs
corpus = []

def process(data):
    for l in data.split("===="):
        if l != "\n":
            row = l.split(">")
            corpus.append(row[0].lower())
            

#make a corpus of documents based on emails' text
data = open("raw_emails_ip.txt").read()
process(data)

#bag of words
bag_of_words = []

#stop words
temp = open("stop_words").readlines()
stop_words = [w.replace("\n","") for w in temp]

#read each doc, tokenize and put it in bag of words
import re
for d in corpus:
    for t in re.split(r'\W+',d): #split by more than 1 space
        for w in re.split(r"[^a-zA-Z0-9]", t): #if any word has special characters, split again
            if w not in stop_words and len(w) > 2:
                bag_of_words.append(w)

#make a set i.e. unique list of words i.e. final bag of words
bag_of_words = list(set(bag_of_words))
#print bag_of_words

#make feature => index  map, which gives the index of value of the feature in vector
dimensions_index_map = {}
for i, w in enumerate(bag_of_words):
    dimensions_index_map[w] = i
#print dimensions_index_map

#make document vectors corpus
vector_corpus = []
for d in corpus:
    v1 = [0] * len(bag_of_words)
    tokens = d.split()
    for w in bag_of_words:
        if v1[dimensions_index_map[w]] == 0:
            cnt = tokens.count(w)
            v1[dimensions_index_map[w]] = cnt
    vector_corpus.append(v1)
        

print len(bag_of_words)
print vector_corpus
        
    
```
