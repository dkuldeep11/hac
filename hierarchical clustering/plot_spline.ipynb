```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline


def plot_curve(x, y1, y2):
    x = np.array(x)
    y1 = np.array(y1)
    y2 = np.array(y2)
    
    x_smooth = np.linspace(x.min(), x.max(), 300)
    y1_smooth = spline(x, y1, x_smooth)
    y2_smooth = spline(x, y2, x_smooth)

    plt.plot(x_smooth,y1_smooth)
    plt.plot(x_smooth,y2_smooth)
    plt.show()
    
    
x = [1,2,3,4]
y = [1,4,9,16]
y1 = [1,8,27,64]
plot_curve(x,y,y1)
```

```python
#fibonnaci naive
def fib(n):
    if (n <= 2):
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    

#fibonacchi DP
memo = {}
def fib_dp((n):
    if n in memo:
        return memo[n]
    else:
        if n <= 2:
            return 1
        else:
            f = fib_dp(n-1) + fib_dp(n-2)
            memo[n] = f
            return f


        
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
y1 = [1,8,27,64]
plot_curve(x,y,y1)
```
